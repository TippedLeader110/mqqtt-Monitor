```python
## program subscriber untuk mqtt
import pymongo
import paho.mqtt.client as mqtt
import json
import datetime
import pytz
import json
import random

myclient = pymongo.MongoClient("mongodb://localhost:27017/")

dbName = "mqtt"

mydb = myclient[dbName]
mycol = mydb["listrik"]

# fungsi yang dipanggil pertama untuk melakukan subscribe ke mqtt broker
def on_subscribe(client, userdata, mid, granted_qos):
    print("Subscribed: "+str(mid)+" "+str(granted_qos))

# fungsi yang bertujuan untuk menampilkan data yang berhasil diterima subscriber
def on_message(client, userdata, msg):
#     print(str(msg.payload))
    payload = json.loads(msg.payload)
    print(payload)
    proc(payload)
    

def proc(payload):
    if mycol.count_documents({ 'time': payload['time'] }, limit = 1) != 0:
        query = {}
        query["time"] = payload['time']
        projection = {}
        projection["listrik"] = u"$listrik"
        projection["_id"] = 0
        cursor = mycol.find(query, projection = projection)
#         print(cursor[0]['listrik'])
        listk = int(payload['listrik']) + int(cursor[0]['listrik'])
        print("Mempebaharui data menjadi  = ",listk)
        myquery = { "time": payload['time'] }
        newvalues = { "$set": { "listrik": listk } }
        mycol.update_one(myquery, newvalues)
    else:
        print('Data tidak ada. Menambah data ....')
        x = mycol.insert_one(payload)

client = mqtt.Client()
client.on_subscribe = on_subscribe
client.on_message = on_message


client.connect("broker.hivemq.com", 1883, 60)
client.subscribe("/monHome/listrik",qos=1)
client.loop_forever()
```
